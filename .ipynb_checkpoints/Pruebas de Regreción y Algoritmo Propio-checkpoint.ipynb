# Modelo

Como al principio dividimos en caracteristicas categoricas y numerocas, ahora es tiempo de juntarlas para crear el data set final de entrenamiento.

```python
# IMPORT
from sklearn.metrics import make_scorer
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV

from sklearn.gaussian_process import GaussianProcessRegressor
from numpy.linalg import inv
from numpy import matmul, transpose
```

```python
train = pd.concat([train_cat,train_num],axis=1)
train.shape
```

Ahora para la etapa de validacion necesitamos separar el data set de entrenamiento para realizar una porterior verificación.

```python
ds_train.SalePrice = np.log1p(ds_train.SalePrice )
y = ds_train.SalePrice
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,y,test_size = 0.3,random_state= 0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train.head(3)
```

Para realizar la separacion de test, usamos K-Fold, con k = 5 iteraciones

```python
n_folds = 5
```

```python
scorer = make_scorer(mean_squared_error, greater_is_better = False)

def rmse_CV_train(model):
    kf = KFold(n_folds,shuffle=True,random_state=42).get_n_splits(train.values)
    rmse = np.sqrt(-cross_val_score(model,X_train,y_train,scoring ="neg_mean_squared_error",cv=kf))
    return (rmse)

def rmse_CV_test(model):
    kf = KFold(n_folds,shuffle=True,random_state=42).get_n_splits(train.values)
    rmse = np.sqrt(-cross_val_score(model,X_test,y_test,scoring ="neg_mean_squared_error",cv=kf))
    return (rmse)
```

# Aplicamos regreción Linear

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
test_pre = lr.predict(X_test)
train_pre = lr.predict(X_train)
print('rmse en train',rmse_CV_train(lr).mean())
print('rmse en train',rmse_CV_test(lr).mean())
```

```python
#plot between predicted values and residuals
plt.scatter(train_pre, train_pre - y_train, c = "orange",  label = "Datos de entrenamiento")
plt.scatter(test_pre,test_pre - y_test, c = "green",  label = "Datos de Validación")
plt.title("Regresión Lineal")
plt.xlabel("Valores predecidos")
plt.ylabel("Reciduos")
plt.legend(loc = "upper left")
plt.hlines(y = 0, xmin = 10.5, xmax = 13.5, color = "red")
plt.show()
```

```python
# Plot predictions - Real values
plt.scatter(train_pre, y_train, c = "orange",  label = "Datos de entrenamiento")
plt.scatter(test_pre, y_test, c = "green",  label = "Datos de validación")
plt.title("Regresión Lineal")
plt.xlabel("Valores Predecidos")
plt.ylabel("Valores reales")
plt.legend(loc = "upper left")
plt.plot([10.5, 13.5], [10.5, 13.5], c = "red")
plt.show()
```
